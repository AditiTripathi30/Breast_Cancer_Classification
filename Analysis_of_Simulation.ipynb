```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file1 = ('/content/drive/MyDrive/rmsd_w.xvg')
data1 = np.loadtxt(file1, skiprows=18)
file2 = ('/content/drive/MyDrive/rmsd.xvg')
data2 = np.loadtxt(file2, skiprows=18)
```

```python
time1 = data1[:, 0]
rmsd1 = data1[:, 1]

time2 = data2[:, 0]
rmsd2 = data2[:, 1]
```

```python
csv_file_path = ('/content/drive/MyDrive/Residues.csv')
csv_data = pd.read_csv(csv_file_path)
x_values = csv_data
```

```python
xvg_file_path1 = ('/content/drive/MyDrive/rmsf_rna_m.xvg')
with open(xvg_file_path1, 'r') as xvg_file:
    lines = xvg_file.readlines()

rmsf_values1 = []
for line in lines:
    if line.startswith('@') or line.startswith('#'):
        continue
    tokens = line.strip().split()
    if len(tokens) == 2:
        rmsf_values1.append(float(tokens[1]))

xvg_file_path2 = ('/content/drive/MyDrive/rmsf_rna.xvg')
with open(xvg_file_path2, 'r') as xvg_file:
    lines = xvg_file.readlines()

rmsf_values2 = []
for line in lines:
    if line.startswith('@') or line.startswith('#'):
        continue
    tokens = line.strip().split()
    if len(tokens) == 2:
        rmsf_values2.append(float(tokens[1]))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_values, rmsf_values1, label='Wildtype')
plt.plot(x_values, rmsf_values2, label='R128A')
plt.xlabel('Residue Index')
plt.ylabel('RMSF (nm)')
plt.title('RMSF RNA Comparison')
plt.legend()
plt.show()
```
